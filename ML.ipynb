#Q-11
How do you approach a Machine Learning problem?  
--> The general approach to solving a Machine Learning problem is:  
1. Define the problem and objective.  
2. Collect relevant data.  
3. Clean and preprocess the data.

#Q-16

 What is an Optimizer? What are different types of optimizers? Explain each with an example.  
--> An optimizer is an algorithm that updates model parameters (weights and biases) to minimize the loss function.  
--> Common types:  
    1. Gradient Descent → Updates weights in the direction opposite to the gradient.  
       Example: w = w - α * dw  
    2. Stochastic Gradient Descent (SGD) → Updates weights for each data sample instead of the whole dataset.  
    3. Adam Optimizer → Combines RMSprop and Momentum; adapts learning rate for each parameter.  
       Example: from tensorflow.keras.optimizers import Adam

#Q-21 What is feature scaling? How does it help in Machine Learning?  
--> Feature scaling standardizes or normalizes data so that all features contribute equally to model performance.  
--> Helps algorithms like Gradient Descent converge faster and prevents bias due to large-scale features.  
   Example: StandardScaler() or MinMaxScaler()  

----------------------------------------------------------------------------------------------

#Q-25 Explain data encoding.  
--> Data encoding converts categorical data into numeric form for model training.  
--> Common types:  
    1. Label Encoding → Assigns each category an integer value.  
       Example:  
       from sklearn.preprocessing import LabelEncoder  
       le = LabelEncoder()  
       X['Gender'] = le.fit_transform(X['Gender'])  
    2. One-Hot Encoding → Creates binary columns for each category.  
       Example:  
       pd.get_dummies(X, columns=['City'])
